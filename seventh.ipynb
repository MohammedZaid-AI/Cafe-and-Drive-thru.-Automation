```python
import csv
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("EmployeeText.csv")
```

```python
data.columns=data.columns.str.strip()
data['Salary'] =pd.to_numeric(data["Salary"].astype(str).str.replace(',',''),errors='coerce')
data["Experience"]=pd.to_numeric(data['Experience'],errors='coerce')
print(data.dtypes)
```

```python
data['Bonus']=np.where(data['Experience']> 5, 0.10*data["Salary"],0)
# data.sort_values(by=["Bonus"],ascending=False )
data = data.sort_values(by="Bonus",ascending=False)
```

```python
data.to_csv("EmployeeText_new.csv",index="False")
```
